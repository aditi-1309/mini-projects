```python
#Intro to Seaborn
import matplotlib.pyplot as plt
import seaborn as sns #Library based on matplotlib
%matplotlib inline
plt.rcParams['figure.figsize']=8,6
import warnings
warnings.filterwarnings('ignore')
```

```python
#Distribution
vis1 = sns.distplot(stats.InternetUsers)
```

```python
vis1 = sns.distplot(stats.InternetUsers, bins=20)
```

```python
#Boxplot
vis2 = sns.boxplot(data=stats,x="IncomeGroup", y="BirthRate")
#Use Seaborn gallery for various types of plots - https://seaborn.pydata.org/examples/index.html#
```

```python
#KEYWORD ARGUMENTS
#Add colors. plt.scatter is used in lmplot. How to use it right
vis3 = sns.lmplot(data=stats, x = 'InternetUsers',y='BirthRate',fit_reg=False, hue='IncomeGroup',size = 8, \
                  scatter_kws={"s":100})
```
